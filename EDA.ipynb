# Exploratory Data Analysis on E-Commerce Cosmetic Products Dataset


### Project Overview:
This project focuses on performing Exploratory Data Analysis (EDA) on an e-commerce cosmetics dataset to extract meaningful business insights about product trends, pricing, and customer preferences. The dataset includes various cosmetic product listings with attributes such as product name, price, size, rating, brand, color, and more

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Downloads/E-commerce  cosmetic dataset.csv/E-commerce  cosmetic dataset.csv',encoding='latin1')
df.info()
```

```python
df.describe()
df.isnull().sum()
```

## Data Cleaning

### Converting Data Type

```python
df['size'] = pd.to_numeric(df['size'],errors = 'coerce')
df['rating'] = pd.to_numeric(df['rating'],errors = 'coerce')
df['noofratings'] = pd.to_numeric(df['noofratings'],errors = 'coerce')
```

### Handling Missing Values

```python
df['price'].fillna(df['price'].mean(),inplace=True)
df['size'].fillna(df['size'].median(),inplace=True)
df['rating'].fillna(df['rating'].median(),inplace=True)
df['noofratings'].fillna(df['noofratings'].mean(),inplace=True)
```

```python
df.dropna(axis=0,inplace=True)
```

## Data Visualization

### Distribution of Price

This plot shows the distribution of product prices in the dataset, helping us understand typical pricing, the presence of outliers, and whether the data is skewed.

```python
sns.displot(df['price'],bins= 20,color='green',kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price (in INR)')
plt.ylabel('Number of Products')
plt.tight_layout()
plt.show()
```

Most products are priced between 1000 and 5000, with a long right tail indicating a few high-priced items.

### Product Form


This chart displays the frequency of each product form , revealing which forms are most common in the inventory.

```python
sns.countplot(x='form',data = df,color='darkorchid')
plt.xticks(rotation=45)
plt.title('Number of Product Form')
plt.xlabel('Form')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

Liquid is the most prevalent form


### Rating Distribution

This chart compares the distribution of product ratings across different categories, highlighting leaders and laggards in customer satisfaction.


```python
sns.boxplot(x='category', y='rating',hue='category',legend=False, data = df,palette='Dark2')
plt.title("Rating by Product Category")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.tight_layout()
plt.show()
```

### Price vs Rating

This scatter plot explores the relationship between product price and customer rating, revealing potential trends or clusters.


```python
sns.scatterplot(x='price',y='rating',data=df,hue='category',size='category',sizes=(10,100))
plt.xlim(0,5000)
plt.title("Price vs Rating by Category")
plt.xlabel('Price')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

There is no strong correlation between price and rating both high and low-priced products receive a range of ratings.

### Number of Products Ordered 

This chart compares product prices across different countries, revealing regional pricing strategies or market differences.


```python
sns.barplot(x='category',y='price',data=df,hue='country',palette='rocket_r')
plt.title('Price by Country')
plt.xlabel('Category')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

This chart reveals that while the USA generally has higher prices for most beauty product categories, certain segments (eyes, face, lips) are more expensive in India. These differences may reflect market positioning, consumer preferences, or supply chain factors in each country

# Recommendations

2. Review Underperformers: Investigate and consider discontinuing products with consistently low sales or poor ratings.

3. Optimize Pricing: Adjust pricing strategies based on category and regional trends to stay competitive and boost profits.
